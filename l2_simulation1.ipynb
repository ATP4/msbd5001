```python
%matplotlib inline
# This is a simulation that demonstrates the difference between kNN and Linear Models
# Case 2: If the data can be viewed as mixture of low variance distributions
# Please refer to chapter 2 of reference book ESLII for details
```

```python
import matplotlib.pyplot as plt
import sklearn
import numpy as np
import random
random.seed(5001)
```

```python
mean_pos = [2,0]
cov_pos = [[3,1.5],[1.5,3]]
x_pos = np.random.multivariate_normal(mean_pos, cov_pos, 100)
xs_pos = [i[0] for i in x_pos]
ys_pos = [i[1] for i in x_pos]


mean_neg = [-2,0]
cov_neg = [[2,1],[1,2]]
x_neg = np.random.multivariate_normal(mean_neg, cov_neg, 100)
xs_neg = [i[0] for i in x_neg]
ys_neg = [i[1] for i in x_neg]
```

```python
plt.figure(dpi=300)
plt.xlim(-8,8)
plt.ylim(-8,8)
dot_size = 6
plt.scatter(xs_pos, ys_pos, color='blue',s=dot_size)
plt.scatter(xs_neg, ys_neg, color='red',s=dot_size)
```

```python
data = np.concatenate((x_pos,x_neg),axis=0)
label = []
for _ in range(100):
    label.append(1)
for _ in range(100):
    label.append(0)
```

```python
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier(loss="log",tol=0.001)
clf.fit(data,label)
```

```python
from sklearn.neighbors import KNeighborsClassifier
n_clf = KNeighborsClassifier(n_neighbors=5)
n_clf.fit(data,label)

X=data
y=label
h=0.02
x_min, x_max = -8,8#X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = -8,8#X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

from matplotlib.colors import ListedColormap
# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
```

```python
Z = n_clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(dpi=300)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, s=6)
```

```python
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(dpi=300)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, s=6)
```
